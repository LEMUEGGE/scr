```python
%matplotlib inline
def ran_sign(size):
    ran = np.random.rand(size)
    out = [1 if i < 0.5 else -1 for i in ran]
    return out
N = 10000
sampling = 7
sensor_array = []
setpoint_array = []
label_array = []
start_ar = np.random.randint(0,100,N)
mag_ar = np.random.randint(-50,50,N)
print(mag_ar)
startval_ar = np.random.randint(80,130,N)
len1_ar = np.random.randint(2,10,N)
len2_ar = np.random.randint(10,15,N)
offset_ar = np.random.randint(0,7,N)
for start,mag,startval,len1,len2,offset in zip(*[start_ar,mag_ar,startval_ar,len1_ar,len2_ar,offset_ar]):
    setpoint = np.ones(150)*startval
    setpoint[start:] = startval+mag
    sensor_value2 = setpoint.copy() + np.random.randint(1,8,1)*np.random.rand(150)*ran_sign(150)
    sensor_value2[start+offset:start+len1+offset] = [sensor_value2[start-1]+(setpoint[start+len1]-setpoint[start-1])/len1*i for i in np.arange(len1)]
    if offset > 0:
        sensor_value2[start:start+offset] = setpoint[start-1] + np.random.randint(1,8,1)*np.random.rand(offset)*ran_sign(offset)
    add = (-np.arange(len2)/len2*0.2+1)*np.sin(np.arange(len2)*0.5)*abs(mag)/np.random.randint(1,10) + np.random.rand(len2)*ran_sign(len2) *3
    sensor_value2[start+len1+offset:start+len1+len2+offset] = setpoint[start+len1]+add if mag > 0 else setpoint[start+len1]- add
    label = []
    
    weird_offset = True if np.random.rand(1)>1 else False
    if not weird_offset:
        for i in np.arange(0,150,sampling):
            if i in np.arange(start-sampling,start+len1+len2+offset):
                label.append(1)
            else:
                label.append(0)
    else:
        sensor_value2[start+len1+len2+offset:] = sensor_value2[start+len1+len2+offset:] + (1.5 + np.random.rand(1)*2.5)*ran_sign(1)
        for i in np.arange(0,150,sampling):
            if i in np.arange(start-sampling,150):
                label.append(1)
            else:
                label.append(0)
        if len(label) != 30:
                print("test")
                break
    """plt.plot(setpoint)
    plt.plot(sensor_value2)
    plt.plot(np.array([[i]*5 for i in label]).flatten())
    plt.show()"""
    sensor_value2 = np.array(sensor_value2).flatten()
    setpoint_array.append(setpoint)
    sensor_array.append(sensor_value2)
    label_array.append(label)
sensor_value2.shape
```
